```python
import random

# Word list
words = ["house", "computer", "apple", "dog", "cat", "university", "library", "friendship", "airport", "garden", "slrg", "clown", "ball", "pool", "river"]
secret_word = random.choice(words).lower()

# ASCII art for each stage of the hangman
hangman_stages = [
    """
       ------
       |    
       |    
       |    
       |    
       |
    --------
    """,
    """
       ------
       |    |
       |    
       |    
       |    
       |
    --------
    """,
    """
       ------
       |    |
       |    O
       |    
       |    
       |
    --------
    """,
    """
       ------
       |    |
       |    O
       |    |
       |    
       |
    --------
    """,
    """
       ------
       |    |
       |    O
       |   /|
       |    
       |
    --------
    """,
    """
       ------
       |    |
       |    O
       |   /|\\
       |    
       |
    --------
    """,
    """
       ------
       |    |
       |    O
       |   /|\\
       |   /
       |
    --------
    """,
    """
       ------
       |    |
       |    O
       |   /|\\
       |   / \\
       |
    --------
    """,
]

# Game variables
max_attempts = len(hangman_stages) - 1  # Maximum number of wrong attempts
guessed_letters = []
correct_letters = ["_"] * len(secret_word)  # Progress on the word
attempts = 0  # Wrong attempts counter

print("Welcome to Hangman!")

# Main game loop
while attempts < max_attempts:
    # Display current hangman stage and word progress
    print(hangman_stages[attempts])
    print("Current word: ", " ".join(correct_letters))
    print("Guessed letters: ", " ".join(guessed_letters))
    
    # Get the player's guess
    guess = input("Guess a letter: ").lower()
    
    # Input validation
    if len(guess) != 1 or not guess.isalpha():
        print("Please enter a single alphabetical letter.")
        continue
    if guess in guessed_letters:
        print("You've already guessed that letter!")
        continue
    
    # Add the guess to guessed letters
    guessed_letters.append(guess)
    
    # Check if the guessed letter is in the secret word
    if guess in secret_word:
        print(f"Good job! '{guess}' is in the word.")
        # Reveal the correct letters in the word
        for idx, letter in enumerate(secret_word):
            if letter == guess:
                correct_letters[idx] = guess
        # Check if the player has guessed the entire word
        if "_" not in correct_letters:
            print("Congratulations! You guessed the word:", secret_word)
            break
    else:
        # Wrong guess, increment attempts
        print(f"Sorry, '{guess}' is not in the word.")
        attempts += 1
    
    # Check if the player has run out of attempts
    if attempts == max_attempts:
        print(hangman_stages[attempts])  # Final hangman stage
        print("Game over! You've been hanged.")
        print("The word was:", secret_word)
        break
```
